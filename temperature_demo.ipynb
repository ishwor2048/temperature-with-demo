This notebook demonstrates how to create a basic chatbot using OpenAI's GPT-4o-mini model, deploy it with Gradio, and allow users to adjust the `temperature` parameter via a slider.

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

# Initialize the LLM model with a default temperature
# GPT-4o-mini is a good choice for cost-effective and capable chatbot applications
def get_chatbot_response(user_message: str, temperature: float) -> str:
    """
    Generates a response from the GPT-4o-mini chatbot.

    Args:
        user_message: The user's input message.
        temperature: The creativity parameter for the LLM. Higher values (e.g., 0.7-1.0)
                     make the output more random, while lower values (e.g., 0.0-0.3)
                     make it more focused and deterministic.

    Returns:
        The chatbot's generated response as a string.
    """
    chat = ChatOpenAI(model="gpt-4o-mini", temperature=temperature)

    messages = [
        SystemMessage(content="You are a helpful and friendly AI assistant. Answer the user's questions truthfully and concisely."),
        HumanMessage(content=user_message),
    ]

    response = chat.invoke(messages)
    return response.content

print("Chatbot function defined.")
```

### Deploying the Chatbot with Gradio

Gradio provides an easy way to create customizable web interfaces for machine learning models. Here, we'll create an interface with an input text box for the user message, a slider for the `temperature` parameter, and an output text box for the chatbot's response.

```python
import gradio as gr

# Create the Gradio interface using gr.Blocks for custom layout
with gr.Blocks(title="GPT-4o-mini Chatbot with Temperature Control") as demo:
    gr.Markdown(
        """
        # GPT-4o-mini Chatbot with Temperature Control
        Interact with a GPT-4o-mini powered chatbot and adjust its creativity using the temperature slider.
        """
    )

    # Place message input and temperature slider vertically
    message_input = gr.Textbox(lines=2, placeholder="Enter your message here...", label="Your Message")
    temperature_slider = gr.Slider(minimum=0.0, maximum=1.5, step=0.1, value=0.7, label="Temperature (0.0 = more predictable, 1.0 = more creative)")

    # Add a Send button
    send_button = gr.Button("Send")

    response_output = gr.Textbox(label="Chatbot Response", interactive=False)

    # Connect the inputs and outputs to the chatbot function
    # The send_button click event will trigger the chatbot function
    send_button.click(
        fn=get_chatbot_response,
        inputs=[message_input, temperature_slider],
        outputs=response_output
    )
    # Also keep the message_input.submit for convenience (Enter key)
    message_input.submit(
        fn=get_chatbot_response,
        inputs=[message_input, temperature_slider],
        outputs=response_output
    )

    # Add a clear button
    gr.ClearButton(components=[message_input, response_output])

# Launch the Gradio app
demo.launch(share=True)
```
